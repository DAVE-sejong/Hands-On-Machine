머신러닝은 데이터로부터 학습하도록 컴퓨터를 프로그래밍하는 기술입니다.

대용량의 데이터를 분석하면 겉으로는 보이지 않았던 패턴을 발견할 수 있는 기술을 데이터 마이닝이라고 합니다.

* 머신러닝 시스템의 종류

지도학습에는 알고리즘에 주입하는 훈련 데이터에 레이블이라는 원하는 답이 포함됩니다.

예를 들면, 분류나 타깃 수치를 예측하는 것입니다.

비지도 학습에는 레이블이 없습니다. 

예를 들면, 계층 군집 알고리즘, 시각화 알고리즘, 이상치 탐지, 연관 규칙 학습 등을 예로 들수 있습니다.

준지도 학습은 레이블이 일부만 존재하는 데이터를 학습하는 알고리즘입니다.

강화학습은 에이전트(environment)를 관찰해서 행동을 실행하고 보상(reward)이나 벌점(penalty)을 받게 됩니다. 시간이 지나면서 큰 보상을 얻기 위해 정책(policy)이라 부르는 최상의 전략을 스스로 학습합니다.

배치학습에서는 시스템이 점진적으로 학습할 수 없습니다.

학습한 것을 단지 적용만 시키는 것를 오프라인 학습이라고 합니다.

온라인 학습에서는 데이터를 순차적으로 한개씩 또는 미니배치라 부르는 작은 묶음 단위로 주입하여 시스템을 훈련시킵니다.

온라인 학습 시스템에서 중요한 파라미터 하나는 변화하는 데이터에 얼마나 빠르게 적응할 것인지 나타내 주는 것을 학습률이라고 부릅니다.

머신러닝 시스템은 어떻게 일반화되는가에 따라 분류할 수도 있습니다.

사례 시반 학습은 사례를 기억함으로써 유사도 측정을 사용해 새로운 데이터에 일반화합니다.

모델 기반 학습은 모델을 만들어 예측에 사용하는 것을 의미합니다.

충분하지 않은 양의 훈련 데이터 때문에 학습이 제대로 안 이루어질 수 있습니다.

대표성 없는 훈련 데이터를 사용하면 여러 문제점이 생기는데, 그 샘플이 작으면 샘플링 잡을이 생기고 매우 큰 샘플도 표본 추출 방법이 잘못 되면 대표성을 띠지 못할 수 있습니다. 이를 샘플링 편향이라고 부릅니다.

낮은 품질의 데이터, 관련 없는 특성이 있는 데이터로 학습을 하면 학습결과가 잘 못 나올 수 있습니다.

훈련데이터 과대적합의 위험을 감소시키기 위해 모델에 제약을 가하는 것을 규제라고 부릅니다.

학습하는 동안 적용할 규제의 양은 하이퍼파라미터(hyperparameter)가 결정합니다.

모델이 새로운 샘플에 얼마나 잘 일반화될지 아는 유일한 방법은 새로운 샘플에 실제로 적용시켜보는 것입니다. 일반적인 해결방법으로는 검증 세트라 부르는 두 번째 홀드아웃 세트를 만드는 것입니다. 예를 들면 교차 검증 기법을 사용합니다.
